Esse dataset apresenta dados sobre os comportamentos de usuários de dispositivos móveis, buscando analisar fatores como o tempo de tela dos usuários, o consumo de bateria e outros dados pessoais. Por fim, gerando um tipo de classificação para esse usuário.

# Importação de dados

```python
import pandas as pd

base_comportamento = pd.read_csv("user_behavior_dataset.csv", delimiter = ",")

df = base_comportamento

display(df)
```

# Classificação dos dados

```python
dados = df.columns.tolist()

display(dados)
```

# Tabelas de frequência - Modelos dos dispositivos

```python
frequencia = df['Device Model'].value_counts()
frequenciaR = df['Device Model'].value_counts(normalize=True)


frequencia = pd.DataFrame(frequencia)
frequenciaR = pd.DataFrame(frequenciaR)


display(frequencia)
display(frequenciaR)
```

# Tabelas de frequência - Gênero

```python
frequenciaR = df['Gender'].value_counts(normalize=True) * 100
frequencia = df['Gender'].value_counts()

frequencia = pd.DataFrame(frequencia)
frequenciaR = pd.DataFrame(frequenciaR)

display(frequencia)
display(frequenciaR)
```

# Tabelas de frequência do tempo de tela dos usuários

```python
frequenciaR = df['Screen On Time (hours/day)'].value_counts(normalize=True) * 100
frequencia = df['Screen On Time (hours/day)'].value_counts()

frequencia = pd.DataFrame(frequencia)
frequenciaR = pd.DataFrame(frequenciaR)

display(frequencia)
display(frequenciaR)
```

# Tabelas de frequência - Idades dos usuários

```python
frequencia = df['Age'].value_counts()
frequenciaAC = df['Age'].value_counts(normalize=True) * 100

frequencia = pd.DataFrame(frequencia)
frequenciaAC = pd.DataFrame(frequenciaAC)


display(frequencia)
display(frequenciaAC)
```

# Gráfico de distribuição de idades dos usuários

```python
import matplotlib.pyplot as plt
%matplotlib inline
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

frequencias = df["Age"].value_counts()

linhas = frequencias.index          # Idades
colunas = frequencias.values        # Frequências correspondentes

# Criando o gráfico de barras
ax.bar(linhas, colunas, color='skyblue', edgecolor='black')

# Configurações do gráfico
ax.set_title('Distribuição de Idades')
ax.set_xlabel('Idade')
ax.set_ylabel('Frequência')
plt.show()
```

# Gráfico de distribuição de modelos dos dispositivos

```python
import matplotlib.pyplot as plt
%matplotlib inline
ax = fig.add_axes([0, 0, 1, 1])

frequencias = df["Device Model"].value_counts()

# Criando o gráfico de pizza
plt.figure(figsize=(8, 8))
plt.pie(
    frequencias.values,                  # Valores das frequências
    labels=frequencias.index,            # Rótulos correspondentes
    autopct='%1.1f%%',                   # Formato das porcentagens
    startangle=140,                      # Ângulo inicial
    colors=plt.cm.tab20.colors           # Paleta de cores
)

plt.title('Distribuição de modelos dos dispositívos')
plt.show()
```

# Gráfico de dispersão do tempo de tela dos usuários

```python
%matplotlib inline
frequencias = df["Screen On Time (hours/day)"].value_counts()

# Criando o gráfico de dispersão
plt.figure(figsize=(10, 6))
plt.scatter(
    frequencias.index,
    frequencias.values,
    color='blue',
    alpha=0.7,
    edgecolor='black'
)

# Configurações do gráfico
plt.title('Dispersão do Tempo de tela dos usuários')
plt.xlabel('Tempo de tela (Horas por dia)')
plt.ylabel('Frequência')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

# Histograma do tempo de uso de aplicativos por dia

```python
%matplotlib inline


plt.figure(figsize=(10, 6))

plt.hist(
    df['App Usage Time (min/day)'],
    bins=50,                  # Número de intervalos
    color='blue',             # Cor do histograma
    edgecolor='black',        # Cor das bordas das barras
    alpha=0.7                 # Transparência das barras
)

# Configurações do gráfico
plt.title('Tempo de uso de aplicativos')
plt.xlabel('Minutos/Dia')
plt.ylabel('Frequência')
plt.grid(linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

# Histograma de consumo de bateria por dia (mAh/Dia)

```python
%matplotlib inline


plt.figure(figsize=(10, 6))

plt.hist(
    df['Battery Drain (mAh/day)'],
    bins=50,
    color='blue',
    edgecolor='black',
    alpha=0.7
)

# Configurações do gráfico
plt.title('Consumo de Bateria')
plt.xlabel('mAh/Dia')
plt.ylabel('Frequência')
plt.grid(linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

# Média do tempo de tela dos usuários (horas/dia)

# Mediana do tempo de tela dos usuários (horas/dia)

```python
df = pd.read_csv("user_behavior_dataset.csv")
mediana = df["Screen On Time (hours/day)"].median()
print(f"Mediana: {mediana}")
```

# Valores de máximo e mínimo do tempo de tela dos usuários (horas/dia)

```python
df = pd.read_csv("user_behavior_dataset.csv")
maximo = df["Screen On Time (hours/day)"].max()
minimo = df["Screen On Time (hours/day)"].min()


print(f"Máximo: {maximo}")
print(f"Mínimo: {minimo}")
```

# Desvio Padrão (Tempo de tela)

```python
desvio_padrao = df["Screen On Time (hours/day)"].std()
print(f"Desvio padrão: {desvio_padrao}")
```
